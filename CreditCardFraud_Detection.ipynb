```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('creditcard.csv')
df.head()
```

```python
df.shape
```

```python
df.tail()
```

```python
df.info()
```

```python
#checking the number of missing values in each column
df.isnull().sum()
```

```python
#distrubution of legit transaction & fradulent transaction
df['Class'].value_counts()
```

This Dataset is highly unbalanced

0 -->Normal Transaction
1 -->Fradulent Transaction


```python
#separating the data for analysis
legit =df[df.Class==0]
fraud=df[df.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

Building a sample dataset containing similar distribution of normal transactions and fraudlent transactions

fraudlent transaction --> 492

```python
legit_sample=legit.sample(n=492)
```

Concatenate two DataFrames

```python
new_data=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

Split the data into training data & testing Data

```python
X=new_data.drop(columns='Class',axis=1)
Y=new_data['Class']
```

```python

X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Logistic Regression

```python
model=LogisticRegression()
```

```python
#training the logistic Regression model with Training Data
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print("Accuracy on Training data:",training_data_accuracy)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
cm=confusion_matrix(X_train_prediction,Y_train)
plt.show()
```
